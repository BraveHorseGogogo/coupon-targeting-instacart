```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

Coefficient Sign Interpretation:

    (++) user_product_orders_share
The higher the share of orders that included the product, the higher the chance it will be reordered this time.

    (-) user_cart_new_product_share_mean
The bigger the share of new products per cart for the user, the lower the chance the user will reorder the product, since the cart is already full of other things, there's simply less space.

Convert back to probability:
- de-scale feature by multiplying coefficient with feature stdev
- take exponent

Let's do this for user_product_orders_share, and user_cart_new_product_share_mean:

```python
# get std for both features:
std_orders_share = np.std(X_train_val_small.user_product_orders_share)
std_share_new = np.std(X_train_val_small.user_cart_new_product_share_mean)

print('Standard Deviation for user_product_orders_share: %.3f' %(std_orders_share))
print('Standard Deviation for user_cart_new_product_share_mean: %.3f' %(std_share_new))
```

```python
# convert logreg coefficient to descaled coefficient
coef_orders_share = std_orders_share*lr.coef_[0][6]
coef_share_new = std_share_new*lr.coef_[0][0]

print('Descaled Coefficient for user_product_orders_share: %.3f'%(coef_orders_share))
print('Descaled Coefficient for user_cart_new_product_share_mean: %.3f'%(coef_share_new))
```

```python
# get exp
exp_orders_share = np.exp(coef_orders_share)
exp_share_new = np.exp(coef_share_new)

print('Effect on LogOdds for user_product_orders_share: %.3f'%(exp_orders_share))
print('Effect on LogOdds for user_cart_new_product_share_mean: %.3f'%(exp_share_new))
```

The LogOdds show the multiplicative effect of the feature. In this case, if the share of orders containing the product is higher, the chance of the person reordering the product will be higher. Likewise, the higher the share of new products per order, the lower the chance of the person reordering a certain product, since there is simply less space for reorders as opposed to new products in the cart.

```python
p_change_orders = exp_orders_share/(1+exp_orders_share)
print(
    'The probability of someone who has only ever bought a certain product, to purchase it again, all else equal, is: \n%.3f' %(p_change_orders))
```

```python
p_change_new = exp_share_new/(1+exp_share_new)
print(
    'The probability of someone who only ever buys new products, to purchase it again, all else equal, is: \n%.3f' %(p_change_new))
```

Convert back to probability:

Let's do this for user_product_orders_share, and user_cart_new_product_share_mean:

```python
# get std for all four features:
std_orders_share = np.std(X_train_val.user_product_orders_share)
std_share_new = np.std(X_train_val.user_cart_new_product_share_mean)
std_reorders = np.std(X_train_val.user_product_reorders)
std_total_orders = np.std(X_train_val.user_product_total_orders)

print('Standard Deviation for user_product_orders_share: %.3f' %(std_orders_share))
print('Standard Deviation for user_cart_new_product_share_mean: %.3f' %(std_share_new))
print('Standard Deviation for user_product_reorders: %.3f' %(std_reorders))
print('Standard Deviation for user_product_total_orders: %.3f' %(std_total_orders))
```

```python
# convert logreg coefficient to descaled coefficient
coef_orders_share = std_orders_share*lr.coef_[0][-1]
coef_share_new = std_share_new*lr.coef_[0][0]
coef_reorders = std_reorders*lr.coef_[0][-5]
coef_total_orders = std_total_orders*lr.coef_[0][-1]


print('Descaled Coefficient for user_product_orders_share: %.3f'%(coef_orders_share))
print('Descaled Coefficient for user_cart_new_product_share_mean: %.3f'%(coef_share_new))
print('Descaled Coefficient for user_product_reorders: %.3f'%(coef_reorders))
print('Descaled Coefficient for user_product_total_orders: %.3f'%(coef_total_orders))
```

```python
# get exp
exp_orders_share = np.exp(coef_orders_share)
exp_share_new = np.exp(coef_share_new)
exp_reorders = np.exp(coef_reorders)
exp_total_orders = np.exp(coef_total_orders)

print('LogOdds for user_product_orders_share: %.3f'%(exp_orders_share))
print('LogOdds for user_cart_new_product_share_mean: %.3f'%(exp_share_new))
print('LogOdds for user_product_reorders: %.3f'%(exp_reorders))
print('LogOdds for user_product_total_orders: %.3f'%(exp_total_orders))
```

The LogOdds show the multiplicative effect of the feature. In this case, if the share of orders containing the product is higher, the chance of the person reordering the product will be higher. Likewise, the higher the share of new products per order, the lower the chance of the person reordering a certain product, since there is simply less space for reorders as opposed to new products in the cart.
Furthermore, if the number of total orders is higher, this indicates that the user orders frequently. Given that, it si less likely the user will buy a specific product in the next order, since there will be more opportunities to include the product. Most importantly, if a product has been reordered in the past, this causes an exorbitant multiplicative effect on the logodds of buying the product again in th enext order.
